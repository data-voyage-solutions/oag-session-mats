### Reduce size of dataset at load using `chunksize=int`

```python
# Chunk size approach:

df_chunk = pd.read_csv('2016HMDALAR - National.zip', names=col_names, chunksize=1000000)

chunk_list = []  # append each chunk df here 

# Each chunk is in df format
for chunk in df_chunk:  

    # Once the data filtering is done, append the chunk to list
    chunk_list.append(chunk)
    
# concat the list into dataframe 
df_concat = pd.concat(chunk_list)

df_concat.head()
```

```python
# Read in data with pre-set dtypes dictionary
df = pd.read_csv('2016HMDALAR - National.zip', header=None, dtype=column_types)
df.columns = col_names
df.head()
```
